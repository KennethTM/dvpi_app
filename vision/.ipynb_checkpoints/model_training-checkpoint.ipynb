```python
from fastai.vision.all import *
```

```python
data_dir = Path("../data")
img_dir = data_dir/"images_preproc"
model_dir = data_dir/"model"
```

```python
print(len(get_image_files(img_dir)))
```

```python
def get_dls(bs, size):
    dl = ImageDataLoaders.from_folder(
        path=img_dir, 
        valid_pct=0.2, 
        bs=bs, 
        item_tfms=Resize(460),
        batch_tfms=aug_transforms(size=size, min_scale=0.85))
    return(dl)
```

```python
get_dls(128, 112).show_batch()
```

```python
dls = get_dls(128, 112)
learn = cnn_learner(dls, 
                    resnet18,
                    loss_func=CrossEntropyLossFlat(),
                    metrics=[accuracy, top_k_accuracy])
```

```python
learn.fit_one_cycle(10, 3e-3)
```

```python
learn.dls = get_dls(64, 224)
learn.loss_func = LabelSmoothingCrossEntropy()
learn.fit_one_cycle(10, slice(1e-6,1e-4))
```

```python
learn.show_results()
```
